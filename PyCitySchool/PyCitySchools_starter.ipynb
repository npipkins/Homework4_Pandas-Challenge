```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#DataFrame
district_summary = pd.DataFrame()
#Total Number of Schools
district_summary['Total Schools'] = [len(school_data_complete['school_name'].unique())]
#Total Number of Students
Total_Students = len(school_data_complete['Student ID'].unique())
district_summary['Total Students'] = [Total_Students]
#Total Budget
district_summary['Total Budget'] = [sum(school_data_complete['budget'].unique())]
#Average Math Score
district_summary['Average Math Score'] = [(school_data_complete['math_score'].mean())]
#Average Reading Score
district_summary['Average Reading Score'] = [(school_data_complete['reading_score'].mean())]
#Percetnage of Students Passing Math
district_summary['% Passing Math'] = [((len(school_data_complete[school_data_complete['math_score'] >= 70])/Total_Students)*100.0)]
#Percetnage of Students Passing Reading
district_summary['% Passing Reading'] = [((len(school_data_complete[school_data_complete['reading_score'] >= 70])/Total_Students)*100.0)]
#Percetnage of Students Passing Overall
district_summary['% Overall Passing'] = [((len(school_data_complete[(school_data_complete['math_score'] >= 70) 
                                                                             & (school_data_complete['reading_score'] >= 70)])/Total_Students)*100.0)]
#Formatting the Data
district_summary.style.format({'Total Schools': "{:}",
                 'Total Students': "{:,.0f}",
                 'Total Budget': "${:,.2f}", 
                 'Average Math Score': "{:.6f}",
                 'Average Reading Score': "{:.6f}",
                 '% Passing Math': "{:.6f}",
                 '% Passing Reading': "{:.6f}",
                 '% Overall Passing': "{:.6f}"})
```

```python
school = school_data.groupby('school_name')['school_name'].unique()

school_type = school_data.groupby('school_name')['type'].unique()

total_students_per_school = school_data.groupby('school_name')['size'].unique()

budget_per_school = school_data.groupby('school_name')['budget'].unique()

budget_per_student = budget_per_school/total_students_per_school

avg_math_score_per_school = student_data.groupby('school_name')['math_score'].mean()

avg_reading_score_per_school = student_data.groupby('school_name')['reading_score'].mean()

passing_score_math = student_data[student_data['math_score'] >= 70]
passingmath_score_per_school = passing_score_math.groupby('school_name')['math_score'].count()
avg_passing_math_per_school = passingmath_score_per_school/total_students_per_school

passing_score_reading = student_data[student_data['reading_score'] >= 70]
passingenglish_score_per_school = passing_score_reading.groupby('school_name')['reading_score'].count()
avg_passing_reading_per_school = passingenglish_score_per_school/total_students_per_school

perc_passing_math = passingmath_score_per_school/total_students_per_school
perc_passing_reading = passingenglish_score_per_school/total_students_per_school

passing_score_overall = student_data[(student_data['math_score'] >= 70) 
                                                  & (student_data['reading_score'] >= 70)]
overall_score_per_school = passing_score_overall.groupby('school_name')['reading_score'].count()
passing_overall_per_school = overall_score_per_school/total_students_per_school


school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_students_per_school,
    "Total School Budget": budget_per_school,
    "Per Student Budget": budget_per_student,
    "Average Math Score": avg_passing_math_per_school,
    "Average Reading Score": avg_passing_reading_per_school,
    '% Passing Math': perc_passing_math,
    '% Passing Reading': perc_passing_reading,
    "% Overall Passing": passing_overall_per_school})

school_summary = school_summary[['School Type',
                                'Total Students',
                                'Total School Budget',
                                'Per Student Budget',
                                'Average Math Score',
                                'Average Reading Score',
                                '% Passing Math',
                                '% Passing Reading',
                                '% Overall Passing']]

school_summary.style.format({'Total Students': '{:,}',
                            "Total School Budget": "${:,}",
                            "Per Student Budget": "${:.2f}",
                            'Average Math Score': "{:.6f}",
                            'Average Reading Score': "{:.6f}",
                            "% Passing Math": "{.6f}",
                            "% Passing Reading": "{.6f}",
                            "% Overall Passing": "{.6f}"})

school_summary

```
